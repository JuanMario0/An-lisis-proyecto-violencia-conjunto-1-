## Modelo a ajustar
$$\mbox{nro_delitos} = \beta_0 + \beta_1\mbox{(nro_mes)}$$

#### Datos entrenamiento

```python
# Definimos la proporción de datos que se utilizarán para el entrenamiento
frac = 0.7

# Seleccionamos de manera aleatoria el 70% de los datos para el entrenamiento
datos_train = df_count.sample(frac=frac,random_state=42)
```

```python
# Mostramos los primeros 10 datos del conjunto de entrenamiento
datos_train.head(10)
```

```python
# Creamos el DataFrame de prueba tomando los datos que no están en el DataFrame de entrenamiento
datos_test = df_count[~df_count.index.isin(datos_train.index)]
```

```python
# Convertimos los valores de las columnas "NUM_MES" y "total_delitos" del conjunto de datos de entrenamiento en arrays numpy
X = np.array(datos_train['NUM_MES'])
Y = np.array(datos_train['total_delitos'])
```

```python

# Se agrega una nueva dimensión al array X para que sea una matriz
# en lugar de un vector
X = X[:, np.newaxis]
```

```python
#Lo mismo pero con la variable Y
Y = Y[:,np.newaxis]
```

### Calculo de coeficiente

$$\left\{
\begin{array}{rcl}
     \displaystyle\hat \beta_0  & = &  \displaystyle\bar y - \hat\beta_1 \bar x
   \\  \displaystyle\hat \beta_1  & = &  (X^{t}X)^{-1}X^{t}y
\end{array}
\right. $$

calculamos el valor del coeficiente de regresión beta1 utilizando la fórmula de la regresión lineal múltiple, donde X es la matriz de características y Y es el vector de valores objetivo. La fórmula utilizada es la siguiente:

beta1 = (X^T * X)^(-1) * X^T * Y

donde:

X^T es la transpuesta de la matriz X
(X^T * X)^(-1) es la inversa de la matriz (X^T * X)
es el operador de multiplicación de matrices
X.T@X es el producto punto de la matriz X transpuesta con la matriz X
X.T@Y es el producto punto de la matriz X transpuesta con el vector Y
np.linalg.inv() es una función de la biblioteca numpy que se utiliza para calcular la inversa de una matriz

```python
beta1 = float(np.linalg.inv(X.T@X)@X.T@Y)
```

```python
beta1
```

```python
# Calcula la media de la variable dependiente Y
# Resta el producto del coeficiente beta1 y la media de la variable independiente X de la media de Y
# Convierte el resultado a un número decimal y lo guarda en la variable beta0
beta0 = float(Y.mean() - beta1*X.mean())

# Imprime el valor de beta0
beta0
```

```python
# Generamos un array de valores de 64 a 40, equiespaciados
u = np.linspace(64,40)

# Calculamos la variable dependiente (v) para cada valor de u
v = beta0 + beta1*u
```

```python
# Se realiza un gráfico
# Se grafica la recta de regresión lineal creada previamente utilizando los valores de u como eje x y los valores de v como eje y, se utiliza color rojo para la línea y se le asigna una etiqueta 'recta_regresion'
plt.plot(u,v,'r',label='recta_regresion')

# Se grafican los puntos de entrenamiento utilizando los valores de 'NUM_MES' como eje x y los valores de 'total_delitos' como eje y, se le asigna una etiqueta 'puntos_train'
plt.scatter(datos_train['NUM_MES'], datos_train['total_delitos'],label='puntos_train')

# Se grafican los puntos de prueba utilizando los valores de 'NUM_MES' como eje x y los valores de 'total_delitos' como eje y, se le asigna una etiqueta 'puntos_test'
plt.scatter(datos_test['NUM_MES'], datos_test['total_delitos'],label='puntos_test')

# Se le asigna una etiqueta al eje x
plt.xlabel('NUM_MES')

# Se le asigna una etiqueta al eje y
plt.ylabel('total_delitos')

# Se asigna un título al gráfico
plt.title('Datos')

# Se muestra una leyenda con las etiquetas definidas previamente
plt.legend()

# Se muestra el gráfico
plt.show()
```

### Predicción y Error

```python
def prediccion(val):
    """
    Función que toma un valor de entrada y devuelve la predicción correspondiente
    utilizando los coeficientes beta0 y beta1 del modelo de regresión lineal.
    
    Args:
    - val: Valor numérico que se utilizará como entrada para la predicción.
    
    Returns:
    - float: Valor numérico correspondiente a la predicción para el valor de entrada.
    """
    
    # Se calcula la predicción utilizando la fórmula del modelo de regresión lineal.
    pred = beta0 + beta1*val
    
    # Se devuelve la predicción.
    return pred
```

```python
#predecir el número de delitos a partir del número de mes en el conjunto de datos de prueba datos_test. 
#La función prediccion utiliza los coeficientes beta0 y beta1 que se calcularon previamente a partir del 
#conjunto de datos de entrenamiento para hacer la predicción.
predicho = prediccion(datos_test['NUM_MES'])  
```

```python
# Se crea un diccionario llamado "resultado" con dos claves: "valor_real" y "prediccion"
resultado = {'valor_real': datos_test['total_delitos'], 'prediccion': predicho}

# Se crea un objeto DataFrame de Pandas llamado "R" a partir del diccionario "resultado"
R = pd.DataFrame(data=resultado)

R.head(10)
```

```python

# Se define una función llamada "error" que recibe dos parámetros: "real" y "pred"
def error(real,pred):
    return np.sum((real-pred)**2)/len(real) # La función calcula el error cuadrático medio (MSE) entre los valores reales y las predicciones
```

```python
# Se llama a la función "error" con dos parámetros: la columna "valor_real" del DataFrame "R" y la columna "prediccion" del mismo DataFrame.
error_cuadratico = error(R['valor_real'], R['prediccion'])
```

```python
# Imprime el valor del error cuadrático medio (ECM) con un formato específico utilizando la función "format()"
print('ECM = {:3f}'.format(error_cuadratico))
```

```python
# Se define una función llamada "lin_regplot" que toma tres parámetros: las características "X", las respuestas "y" y un modelo de regresión "model"
def lin_regplot(X, y, model):
    # Se crea un gráfico de dispersión utilizando los valores de las características "X" y las respuestas "y"
    plt.scatter(X, y, c='blue')
    # Se traza una línea de regresión utilizando los valores de las características "X" y las respuestas predichas por el modelo "model"
    plt.plot(X, model.predict(X), color='red') 
    # La función no devuelve ningún valor, solo muestra el gráfico
    return None
```

```python
lin_regplot(X, Y, modelo)
plt.title('Modelo de regresión')
plt.xlabel('NUM_MES')
plt.ylabel('total_delitos')
plt.show()
```
